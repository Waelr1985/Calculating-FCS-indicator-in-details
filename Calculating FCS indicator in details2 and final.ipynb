```python
import pandas as pd
```

```python
# total sample
FCS = pd.read_excel('c:/Users/LENOVO/Desktop/FCS.xlsx')
```

```python
FCS['FCG']= FCS['Bread_Rice_Bulgar_Pasta']*2 + FCS['Lentils_Nuts_beans']*3 + FCS['Vegetables']*1 + FCS['Fruits']*1 + FCS['Beef_Chicken_lamb_liver_any other meat']*4 + FCS['Milk_yoghurt_cheese_other milk products']*4 + FCS['Oil_Ghee_Butter_oil']*0.5 + FCS['Halva_Jams']*0.5 
```

```python
# List comprehension:
FCS['FCG1'] = [1 if x<=21 else 3 if x>=35.5 else 2 for x in FCS['FCG']]
```

```python
FCS
```

```python
Count = FCS.FCG1.value_counts()
Percent = FCS.FCG1.value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
final = pd.DataFrame({'count': Count, 'per': Percent})
```

```python
final
```
